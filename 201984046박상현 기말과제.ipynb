```python
import pandas as pd
import numpy as np
```

```python
tel = pd.read_csv("data/telco.csv")
tel
```

```python
#Leave	가입해지여부(1이면 해지)
#MemberWeeks	가입기간
#NewContract	가입갱신여부(1이면 갱신)
#IsData	데이터요금제사용여부(1이면 사용)
#DataAmount	데이터사용량(GB)
#Callcenter	고객센터전화량
#CalMins	월평균전화량(분)
#CalCounts	월평균전화수(횟수)
#MonthlyFee	월평균요금
#OverFee	초과요금
#RoamingMin	로밍시간(분)
```

```python
#na값 확인
tel.isnull().sum()
```

```python
tel.info()
```

```python
tel.describe().T
```

```python
#용량 줄이기
tel['Leave'] = tel['Leave'].astype(np.int8)
tel['MemberWeeks'] = tel['MemberWeeks'].astype(np.int16)
tel['NewContract'] = tel['NewContract'].astype(np.int8)
tel['IsData'] = tel['IsData'].astype(np.int8)
tel['Callcenter'] = tel['Callcenter'].astype(np.int8)
tel['CalCounts'] = tel['CalCounts'].astype(np.int16)
tel['DataAmount'] = tel['DataAmount'].astype(np.float16)
tel['CalMins'] = tel['CalMins'].astype(np.float16)
tel['MonthlyFee'] = tel['MonthlyFee'].astype(np.float16)
tel['OverFee'] = tel['OverFee'].astype(np.float16)
tel['RoamingMin'] = tel['RoamingMin'].astype(np.float16)
```

```python
tel.info()
```

```python
tel
```

```python
tel1=tel[['Leave','MemberWeeks','MonthlyFee']]
tel1.head()
```

```python
#가입기간이 제일오래된 기간 찾기
tel1.nlargest(10,'MemberWeeks').head()
```

```python
#위 상태에서 월평균요금이 가장 적은 5개 찾기
tel1.nlargest(10,'MemberWeeks').nsmallest(5,'MonthlyFee')
```

```python
#가입해지여부에따른 가장 요금이 많은것 찾기
(
tel1.groupby('Leave',as_index=False)
.apply(lambda df: df.sort_values(
 'MonthlyFee', ascending=False
 ).head(1)
)
)
```

```python
#고객센터전화량별,가입여부별 최소 월평균전화수
(
tel[['Callcenter','Leave','CalCounts']]
.sort_values(['Callcenter','Leave','CalCounts'],ascending=[False,True,True])
.drop_duplicates(subset=['Callcenter','Leave'])
)
```

```python
#가입여부별 고객센터전화량
import seaborn as sns
fig, ax = plt.subplots(figsize=(15, 10))
top = tel.Callcenter.value_counts().index[:10]
sns.countplot(x='고객센터전화량',
 data=(tel.assign(
 고객센터전화량=tel.Callcenter.where(tel.Callcenter.isin(top)))
 ),
              hue='Leave',
              hue_order=[0, 1],
 )
```

```python
#데이터 사용량과 월평균요금에대한 산포도
fig, ax = plt.subplots(figsize=(8, 8))
tel.plot.scatter(
 x='DataAmount',
 y='MonthlyFee',
 alpha=.1,
 ax=ax
)
```

```python
#가입기간이 100주를 넘고 가입갱신을한 데이터사용량이 3GB초과 5GB미만인것 찾기
t_1 = tel.MemberWeeks > 100
t_2 = tel.NewContract = 1
t_3 = (tel.DataAmount < 5) & (tel.DataAmount > 3)
t_a = t_1 & t_2 & t_3
```

```python
tel[t_a]
```

```python
#고객센터전화량에 따른 가입기간,데이터사용량,초과요금 평균
(
tel
.groupby('Callcenter')
.agg({'MemberWeeks':'mean','DataAmount':'mean','OverFee':'mean'})
)
```

```python
#가입/갱신별 가입기간,데이터사용량,초과요금,로밍시간 평균
(
 tel
 .groupby(['Leave', 'NewContract'])
 [['MemberWeeks', 'DataAmount', 'OverFee', 'RoamingMin']]
 .agg(['mean'])
)
```
